## for loops

```python
for i in range(2, 10):
  for j in range(2, 10):
    print(i, "*", j, "=", i*j)
```

## while loops

```python
i = 2
while i < 10:
  j = 2

  while j < 10:
    j += 1
    print(f"{i} * {j} = {i*j}")
  i += 1
```

```python
i = 2
j = 2
while i < 10:
  print(f"{i}x{j}={i*j}")
  j = j + 1
  if j == 10:
    j = 2
    i = i + 1
```

## exception handling

- runs the code and does not stop even when there is an error, instead it allows the system to move on and just print the error

```python
def add(x, y):
  print(f"{x} + {y} = {x+y}")
  
def sub(x, y):
  print(f"{x} - {y} = {x-y}")

def mul(x, y):
  print(f"{x} * {y} = {x*y}")

def div(x, y):
  print(f"{x} / {y} = {x/y}")
```

```python
#inserting functions in a list
functions = [add, div, sub, mul]
```

```python
# making a for loop to run all the functions at once

a = 3
b = 1


for x in functions:
  x(a, b)
```

```python
# exception handling (since 3 cannot be divided by 0)


a = 3
b = 0


for x in functions:
  try:
    x(a, b)
  except Exception as e:
    print(e.args)
    # prints the error
```
