```python
#how to merge

x = 2
y = 4
led = ["00000"] * 5
display.show(Image("00000:00000:00000:00000:00000"))
display.set_pixel(x, y, 9)

def new_row():
    a = ["0"]*5
    a[random.randint(0,4)] = "9"
    return "".join(a)

while True:
    #display image
    display.show(Image(":".join(led)))
    led.insert(0, new_row())
    led.pop()
    sleep(2000)
    #control movement
    if button_a.was_pressed():
        display.clear()
        x -= 1
        if x < 0:
            x = 0
        display.set_pixel(x, y, 9)
    if button_b.was_pressed():
        display.clear()
        x += 1
        if x > 4:
            x = 4 
        display.set_pixel(x, y, 9)
```
