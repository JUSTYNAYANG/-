#Microbit

- Moon car Move Like Dust Cleaner



```python
from microbit import *
#running the moon car within a space 
#(in white area, does not go out of black line)

pin15.set_pull (pin15.NO_PULL)
pin16.set_pull (pin16.NO_PULL)

while True:
    p15 = pin15.read_digital()
    p16 = pin16.read_digital()
     
     #if the ground is white
    if p16:
        pin8.write_analog(10)
        pin14.write_analog(0)
        pin2.write_analog(200)
        pin13.write_analog(0)
        sleep(1800)
        

    #if the ground is white    
    elif p15:
        pin8.write_analog(200)
        pin14.write_analog(0)
        pin2.write_analog(10)
        pin13.write_analog(0)
        sleep(1800)
       

        
    #if the ground is black (on track)
    else:
       pin8.write_analog (100)
       pin14.write_analog (0)
       pin2.write_analog (100)
       pin13.write_analog(0)
```
