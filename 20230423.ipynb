#recursive function

recursive means that the funtion calls itself while running

##factorial function

```
n! = n * (n - 1) * (n - 2) * ...... * 1
```

````python
#without recursive

def factorial(n):
  final = 1
  for i in range(n):
    final = final* (i+1)
  return final
````

````python
factorial(5)
````

````python
#with recursive

def factorial(n):
  if n == 0:
    return(1)
  else:
    return n*factorial(n-1)
````

````python
factorial(5)
````

````python
factorial(2)
````

```
factorial(2)
  n = 2
  return n*factorial(n-1) = 2*factorial(1)

factorial(1)
  n = 1
  return n*factorial(n-1) = 1*factorial(0)

factorial(0)
  n = 0
  return 1


1*factorial(0) = 1*1 = 1

2*factorial(1)= 2*1 = 2

final output = 2
```

## turtle

### drawing a tree with recursive function

```
import turtle

turtle.penup()
turtle.left(90)
turtle.goto(0,-150)
turtle.pendown()

turtle.speed("fastest")

def tree(size):
  if size > 10:
    turtle.forward(size)
    turtle.right(20)
    tree(size*0.7)
    turtle.left(40)
    tree(size*0.7)
    turtle.right(20)
    turtle.backward(size)
    

tree(100)
```
